```python
import pandas as pd
import os

# 1. 파일 경로 설정 (회사 PC 내 실제 txt 파일 위치 확인)
# 만약 경로가 다르면 수정하세요!
file_path = '../data/raw/L001_2504021029.txt' 
output_path = '../data/processed/smart_meter_data.csv'

# 저장할 폴더가 없으면 생성
os.makedirs(os.path.dirname(output_path), exist_ok=True)

if os.path.exists(file_path):
    # 2. 데이터 불러오기
    df = pd.read_csv(file_path, sep='\s+', header=None)
    df.columns = ['ID', 'Date', 'Time', 'Velocity', 'Flow_Instant', 'Code', 'Total_Flow', 'Pressure']
    
    # 3. 시계열 변환
    df['Timestamp'] = pd.to_datetime(df['Date'].astype(str).str.zfill(6) + 
                                     df['Time'].astype(str).str.zfill(6), 
                                     format='%y%m%d%H%M%S')
    
    # 4. 분석용 컬럼 정리 및 저장
    cols = ['Timestamp', 'Velocity', 'Flow_Instant', 'Pressure', 'Total_Flow', 'ID', 'Code']
    df_final = df[cols]
    df_final.to_csv(output_path, index=False, encoding='utf-8-sig')
    
    print("✅ CSV 파일 생성 성공!")
    print(f"데이터 기간: {df_final['Timestamp'].min()} ~ {df_final['Timestamp'].max()}")
else:
    print(f"❌ 파일을 찾을 수 없습니다: {os.path.abspath(file_path)}")
```
